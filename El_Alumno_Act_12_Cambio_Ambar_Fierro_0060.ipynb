## las variables
1. CP -->  costo del producto
2. CP -->  cantidad de dinero
3. cambio

## formulas
$cambio = CD - CP$

## Diagrama de flujo Lucidchart
<img src="https://raw.githubusercontent.com/a25308051280060-pixel/Calcular-cambio-if-else/refs/heads/main/DF%20if%20else%20calculo2.png" width="350" height="500" alt="cambio de una compra">)


```python
# para cuando es verdadero (si)
# calculando el cambio usando if else
# datos de entrada CD , CP
# mostrar Cambio si es falso
# si es verdadero mostrar No alcanza
CP = float(input("ingre el costo del ´producto:"))
CD = float(input(" ingrese la cantidad del dinero:"))
if CP > CD:
  print("no alcanza")
else:
  cambio = CD - CP
  print("el cambio es:",cambio)
  print("gracias por su compra")
  print("programa realiado por Ambar Fierro NL = 19")
```

```python
# para cuando es falso (no)
# calculando el cambio usando if else
# datos de entrada CD , CP
# mostrar Cambio si es falso
# si es verdadero mostrar No alcanza
CP = float(input("ingre el costo del ´producto:"))
CD = float(input(" ingrese la cantidad del dinero:"))
if CP > CD:
  print("no alcanza")
else:
  cambio = CD - CP
  print("el cambio es:",cambio)
  print("gracias por su compra")
  print("programa realiado por Ambar Fierro NL = 19")
```
